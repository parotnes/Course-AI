Build me a vector of Celsius and a vector of Fahrenheit that will contain 10 dimensions (temp).

Build for me using keras one layer that contains one neuron receives at the input the vector of Celsius and the response of this is the vector of Fahrenheit.

Reduce the number of epochs to 500

Check the error, using the square root of the errors, with a learning step of 0.01



```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from sklearn.metrics import mean_squared_error

# Create a vector of 10 random Celsius temperatures
celsius_vector = np.random.randint(-10, 31, size=10)

# Convert Celsius to Fahrenheit using the formula F = (9/5)C + 32
fahrenheit_vector = (9/5) * celsius_vector + 32

# Build the model
model = Sequential()
model.add(Dense(units=1, input_dim=1))  # One layer with one neuron

# Compile the model with a specified learning rate
optimizer = Adam(learning_rate=0.01)
model.compile(loss='mean_squared_error', optimizer=optimizer)

# Train the model with reduced epochs to 500
history = model.fit(celsius_vector, fahrenheit_vector, epochs=500, verbose=0)

# Test the model with the same Celsius vector
predicted_fahrenheit = model.predict(celsius_vector)

# Calculate the root mean squared error (RMSE)
rmse = np.sqrt(mean_squared_error(fahrenheit_vector, predicted_fahrenheit))

# Print the RMSE
print("Root Mean Squared Error (RMSE):", rmse)
```

Make a prediction for me what the response to the number 100 will be

```python
# Make a prediction for Celsius value 100
new_celsius_value = np.array([100.0])
predicted_fahrenheit_value = model.predict(new_celsius_value)

# Print the result
print("Celsius:", new_celsius_value)
print("Predicted Fahrenheit:", predicted_fahrenheit_value)
```
